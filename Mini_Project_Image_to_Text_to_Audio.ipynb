Step Two: Opening and Processing the Image

```python
from PIL import Image # Opens and Processes the Image

img = Image.open('/content/Picture1.png')
print(img) # prints the image object to confirm it was loaded
```

Step Three: Image to Text and then to Audio


```python
# Image to Text
from gtts import gTTS
result = pytesseract.image_to_string(img)
print(result)

# Text To Audio
tts = gtts.gTTS(result)  # Default Language is English
tts.save("Final_result.mp3")

#  EXTRA INFORMATION : gTTS(result, lang='hi')  # for Hindi
```

Step Four : Playing the Audio File

```python
from IPython.display import Audio  # Plays the MP3 file
Audio('/content/Final_result.mp3', autoplay=True)
```
